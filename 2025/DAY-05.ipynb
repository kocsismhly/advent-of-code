## Setup

```python
INPUT_PATH = 'assets/2025-05.txt'

with open(INPUT_PATH, 'r') as f:
    content = f.read()

ingredients = content.strip().split('\n\n')

fresh_ranges = ingredients[0].split('\n')
available_ingredients = [int(part) for part in ingredients[1].split('\n')]
```

## Part 1

```python
fresh_ranges_parsed = []
for r in fresh_ranges:
    start, end = map(int, r.split('-'))
    fresh_ranges_parsed.append((start, end))

fresh_count = 0
for item in available_ingredients:
    for start, end in fresh_ranges_parsed:
        if start <= item <= end:
            fresh_count += 1
            break

print(f'⭐️ PART 1: {fresh_count} available ingredients are fresh')
```

## Part 2

```python
fresh_ranges_parsed.sort()
merged_ranges = []

for start, end in fresh_ranges_parsed:
    if merged_ranges and start <= merged_ranges[-1][1] + 1:
        # Merge with previous range Merge with previous range if overlapping
        merged_ranges[-1] = (merged_ranges[-1][0], max(merged_ranges[-1][1], end))
    else:
        merged_ranges.append((start, end))

fresh_count = 0
for start, end in merged_ranges:
    fresh_count += (end - start + 1)


print(f'⭐⭐️️ PART 2: {fresh_count} ingredients are fresh')
```
