## Setup

```python
INPUT_PATH = 'assets/2025-06.txt'

with open(INPUT_PATH, 'r') as f:
    lines = [line.rstrip('\n') for line in f.readlines()]
```

## Part 1

```python
split_lines = [line.strip().split() for line in lines]
numbers = [list(map(int, col)) for col in zip(*split_lines[:-1])]  # List of columns
operations = split_lines[-1]

grand_total = 0

for i, o in enumerate(operations):
    if o == '+':
        grand_total += sum(numbers[i])
    elif o == '*':
        result = 1
        for n in numbers[i]:
            result *= n
        grand_total += result

print(f'⭐️ PART 1: {grand_total} is the grand total')
```

## Part 2

```python
# Building columns character by character
max_len = max(len(line) for line in lines)
columns = []

for col_idx in range(max_len):
    col = []
    for line in lines:
        if col_idx < len(line):
            col.append(line[col_idx])
        else:
            col.append(' ')
    columns.append(col)

# Grouping columns into problems (separated by all-space columns)
problems = []
current_problem = []

for col in columns:
    if all(c == ' ' for c in col):
        if current_problem:
            problems.append(current_problem)
            current_problem = []
    else:
        current_problem.append(col)

if current_problem:
    problems.append(current_problem)

grand_total = 0

for problem in reversed(problems):
    operator = problem[0][-1]

    # Building numbers from each column
    numbers = []
    for col in problem:
        digits = [col[i] for i in range(len(col) - 1) if col[i].strip()]  # Last row not needed
        if digits:
            number = int(''.join(digits))
            numbers.append(number)

    # Applying the operation
    result = 0
    if operator == '+':
        result = sum(numbers)
    elif operator == '*':
        result = 1
        for n in numbers:
            result *= n

    grand_total += result

print(f'⭐️⭐️ PART 2: {grand_total} is the actual grand total')
```
