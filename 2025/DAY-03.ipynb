## Setup

```python
INPUT_PATH = 'assets/2025-03.txt'

with open(INPUT_PATH, 'r') as f:
    banks = [list(line.strip()) for line in f if line.strip()]
```

## Part 1

```python
largest_joltages = []

for bank in banks:
    sorted_batteries = sorted(bank, reverse=True)

    largest_battery = sorted_batteries[0]
    largest_battery_pos = bank.index(largest_battery)

    if largest_battery_pos == len(bank) - 1:
        # If largest is last, we use the second largest as largest
        largest_battery = sorted_batteries[1]
        largest_battery_pos = bank.index(largest_battery)

    second_largest_battery = max(bank[largest_battery_pos + 1:])
    combined = largest_battery + second_largest_battery

    largest_joltages.append(int(combined))

print(f'⭐️ PART 1: {sum(largest_joltages)} is the total output joltage')
```

## Part 2

```python
largest_joltages = []

for bank in banks:
    batteries = []
    largest_battery_pos = -1

    for i in range(12):
        find_until = len(bank) - 11 + i  # Limit to leave room for remaining batteries
        search_range = bank[largest_battery_pos + 1:find_until]

        largest_battery = max(search_range)
        largest_battery_pos += search_range.index(largest_battery) + 1

        batteries.append(largest_battery)

    largest_joltages.append(int(''.join(batteries)))

print(f'⭐️ PART 2: {sum(largest_joltages)} is the new total output joltage')
```
