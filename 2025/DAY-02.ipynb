## Setup

```python
INPUT_PATH = 'assets/2025-02.txt'

with open(INPUT_PATH, 'r') as f:
    ranges = f.readline().split(',')
```

## Part 1

```python
invalid_ids = []

for r in ranges:
    start, stop = r.split('-')
    start, stop = map(int, r.split('-'))

    for num in range(start, stop + 1):
        num_str = str(num)

        if len(num_str) % 2 == 1:
            continue

        mid = len(num_str) // 2
        first_half = num_str[:mid]
        second_half = num_str[mid:]

        if first_half == second_half:
            invalid_ids.append(num)

print(f'⭐️ PART 1: {sum(invalid_ids)} is the sum of invalid IDs')
```

## Part 2

```python
invalid_ids = []

for r in ranges:
    start, stop = map(int, r.split('-'))

    for num in range(start, stop + 1):
        num_str = str(num)

        for leng in range(1, len(num_str)):
            if len(num_str) % leng == 0:
                pattern = num_str[:leng]

                if pattern * (len(num_str) // leng) == num_str:
                    invalid_ids.append(num)
                    break

print(f'⭐️ PART 2: {sum(invalid_ids)} is the sum of invalid IDs using the new rules')
```
