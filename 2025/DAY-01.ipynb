## Setup

```python
INPUT_PATH = 'assets/2025-01.txt'

with open(INPUT_PATH, 'r') as f:
    lines = [line.strip() for line in f if line.strip()]

directions = []
numbers = []

for parts in lines:
    directions.append(parts[0])
    numbers.append(int(parts[1:]))
```

## Part 1

```python
current_number = 50
zeros_found = 0

for d, n in zip(directions, numbers):
    n = n % 100

    if d == 'L':
        current_number = (current_number - n) % 100
    elif d == 'R':
        current_number = (current_number + n) % 100

    if current_number == 0:
        zeros_found += 1

print(f'⭐️ PART 1: {zeros_found} is the password to open the door')
```

## Part 2

```python
current_number = 50
zeros_found = 0

for d, n in zip(directions, numbers):
    hundreds = n // 100
    n = n % 100
    zeros_found += hundreds

    prev = current_number

    if d == 'L':
        current_number = (current_number - n) % 100
    elif d == 'R':
        current_number = (current_number + n) % 100

    # Count zero if we landed on it
    if current_number == 0:
        zeros_found += 1
    # Count zero if we touched it (but didnt start or end on 0)
    elif prev != 0 and ((d == 'L' and prev < n) or (d == 'R' and prev + n > 99)):
        zeros_found += 1

print(f'⭐️⭐️ PART 2: {zeros_found} is the actual password to open the door')
```
